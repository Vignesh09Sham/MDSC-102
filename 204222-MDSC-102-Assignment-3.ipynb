```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.special import gamma
```

# 1. Binomial Distribution

```python
def B(n,p):
    # f = lambda r, n, p: (np.math.factorial(n)/(np.math.factorial(r) * np.math.factorial((n-r)))) * ((p)**r) * ((1-p)**(n-r))
    f = lambda r, n, p: np.math.comb(n,r) * ((p)**r) * ((1-p)**(n-r))
    r = np.linspace(0, n, 100,dtype=int)
    y = [f(i,n,p) for i in r]
    plt.title("Binomial Distribution")
    plt.xlabel("x-values")
    plt.ylabel("f(x)")
    plt.plot(r, y)

    n = 20
    p = 0.5
    y = [f(i,n,p) for i in r]
    plt.plot(r, y)

    n = 10
    p = 0.7
    y = [f(i,n,p) for i in r]
    plt.plot(r, y)

    plt.legend(["n = 30, p = 0.2", "n = 20, p = 0.5", "n = 10, p = 0.7"])
    plt.show()

B(30,0.2)

    
```

## 2. Poisson Distribution

```python
def Poi(lam):
    f = lambda x,lam: np.exp(-lam) *((lam**x) / (np.math.factorial(x)))
    x = np.linspace(0, 30, 100, dtype=int)
    y = [f(i,lam) for i in x]
    plt.title("Poisson Distribution")
    plt.xlabel("x-values")
    plt.ylabel("f(x)")
    plt.plot(x, y)

    lam = 0.4
    y = [f(i,lam) for i in x]
    plt.plot(x, y)

    lam = 0.9
    y = [f(i,lam) for i in x]
    plt.plot(x, y)

    plt.legend(["lam = 0.01", "lam = 0.4", "lam = 0.9"])
    plt.show()

Poi(0.01)
```

# 3. Exponential Distribution

```python
def Exp(lam):
    f = lambda x,lam: lam * np.exp((-lam) * x)
    x = np.arange(0, 30, dtype=int)
    y = [f(i,lam) for i in x]
    plt.title("Exponential Distribution")
    plt.xlabel("x-values")
    plt.ylabel("f(x)")
    plt.plot(x, y)

    lam = 0.01
    y = [f(i,lam) for i in x]
    plt.plot(x, y)

    lam = 5
    y = [f(i,lam) for i in x]
    plt.plot(x, y)

    plt.legend(["a = 2", "a = 0.01", "a = 5"])
    plt.show()

Exp(2)
```

# 5. Gamma Distribution

```python
def Gamma(alpha, lam):
    f = lambda x,alpha,lam: ((lam**alpha)/np.math.factorial(alpha)) * x**(lam - 1) * np.exp((-lam) * x)
    x = np.linspace(0, 10, 100)
    y = [f(i, alpha, lam) for i in x]
    plt.title("Gamma Distribution")
    plt.xlabel("x-values")
    plt.ylabel("f(x)")
    plt.plot(x, y)

    alpha = 4
    lam = 2
    y = [f(i, alpha, lam) for i in x]
    plt.plot(x, y)

    alpha = 3
    lam = 1
    y = [f(i, alpha, lam) for i in x]
    plt.plot(x, y)

    plt.legend(["alpha = 5, lam = 3", "alpha = 4, lam = 2", "alpha = 3, lam = 1"])
    plt.show()

Gamma(5,3)
```
