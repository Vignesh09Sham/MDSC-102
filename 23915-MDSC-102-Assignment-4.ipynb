```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
name = ["class", "Alcohol", "Malic acid", "Ash", "Alcalinity of ash", "Magnesium", "Total phenols", "Flavanoids", "Nonflavanoid phenols", "Proanthocyanins", "Color intensity", "Hue", "OD280/OD315 of diluted wines", "Proline"]
df = pd.read_csv("wine.data", names=name)
df
```

```python
sns.scatterplot(x = "Alcohol", y = 'Malic acid', hue='class', data = df)
```

```python
sns.countplot(x = "Alcohol", data = df)
```

```python
def scatter_plots(df, target, variable, count):
    sns.scatterplot(x = target, y = variable, hue='class', data = df, palette="bright")

```

```python
j = 0
plt.figure(figsize=(30,30))
for i in name[2:]:
    j += 1
    plt.subplot(4,3,j)
    scatter_plots(df, "Alcohol", i, j)
    plt.title("Scatter plot for Alcohol and {}" .format(i))

plt.show()
    
```

* In general, class 1 has the highest percentage of alcohol and the majority of the points in class 2 has the lowest percentage of alchol compared to the other classes.

* plot(1,1) : Class 2 seems to have the highest percentage of malic acid and the two other classes have approximately the same percentage of malic acid.

* plot(1,2) : All the classes approximately have the same percentage of ash.

* plot(1,3) : The alkalinity of the ash of class 2 and 3 is somewhat the same. Class 1 comparatively has percentage of alkalinity.

* plot(2,1) : Class 2 has the lowest level of magnesium compared to the classes and class 1 has the highest.

* plot(2,2) : Class 3 has the lowest percentage of phenols and class 1 seemingly has the highest.

* plot(2,3) : Class 3 has the lowest percentage of flavavoids and class 1 has the highest.
