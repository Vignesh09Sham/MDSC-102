```python
log_Magnesium = np.log(df["Magnesium"])
print(log_Magnesium)
print(log_Magnesium.skew())
```

```python
sqrt_Magnesium = np.sqrt(df["Magnesium"])
print(sqrt_Magnesium)
print(sqrt_Magnesium.skew())
```

```python
Magnesium1 = 1 / df["Magnesium"]
print(Magnesium1.skew())
```

```python
Mag_boxcox, mag_lambda = sp.stats.boxcox(df['Magnesium'])
print(mag_lambda)
```

```python
magnesium_boxcox = ((((df['Magnesium'])**mag_lambda) - 1) / mag_lambda)
print(magnesium_boxcox)
print(magnesium_boxcox.skew())
```

```python
plt.figure(figsize=(15,4))
plt.subplot(1,2,1)
sns.histplot(x="Magnesium", data = df, kde = True)
plt.title("Histogram for Magnesium")
plt.subplot(1,2,2)
sns.histplot(x = magnesium_boxcox, data=magnesium_boxcox, kde = True)
plt.title("Histogram for Magnesium_boxcox")

plt.show()
```

```python
fig,axes = plt.subplots(1,2,figsize=(15,4))
sp.stats.probplot(df['Magnesium'], plot = axes[0])
axes[0].set_title("Magnesium qqplot")
sp.stats.probplot(magnesium_boxcox, plot = axes[1])
axes[1].set_title("Magnesium_boxcox qqplot")
plt.show()
```

```python
j = 0
plt.figure(figsize=(20,70))
for i in name[1:]:
    j+=1
    plt.subplot(13,3,j)
    sns.boxplot(y = i,data = df)
    plt.title("Box plot for {}" .format(i))

    j +=1
    plt.subplot(13,3,j)
    sns.histplot(x = i, kde=True, data=df)
    plt.title("Histogram for {}" .format(i)) 

    j +=1
    plt.subplot(13,3,j)
    sp.stats.probplot(df[i], plot=plt) 
    plt.title("Probplot for {}" .format(i)) 
    plt.legend(labels = [f'Skew:{df[i].skew()}'], loc = "lower right")

plt.show()
```

#### Inferences

* Alcohol - For all 3 plots, we can deduce that alcohol is mostly normal.

* Malic acid - It is definitely right skewed since the right tail of the box plot is longer than the left tail. Also, peak of the histogram is mostly on the left and also the skewness value is almost equal to 1.

* Ash - Normal

* Alcalinity of ash - Normal

* Magnesium - It is right skewed since the right tail is longer than the left tail and the peak of the histogram is mostly on the left. Its skewness value is almost equal to 1.

* Total phenols - It is slightly right skewed because the right tail of the boxplot is longer than the left tail. 

* Flavanoids - It is definitely right skewed because the right tail of the boxplot is longer than the left tail.

* Nonflavanoid phenols - It is right skewed because the right tail of the boxplot is longer than the left tail and the peak of the histogram is towards the left.

* Proanthocyanins - It is right skewed because the right tail of the boxplot is longer than the left tail and the peak of the histogram is towards the left.

* Color intensity - It is right skewed because the right tail of the boxplot is longer than the left tail and the peak of the histogram is towards the left.

* Hue - It is right skewed because the peak of the histogram is slightly to the left and the kde drastically decreases to the right.

* OD280/OD315 of diluted wines - It is left skewed because the peak of the histogram is towards the right.

* Proline - It is right skewed because the right tail of the boxplot is longer than the left tail and the peak of the histogram is towards the left.

```python
cols_to_be_normalized= ["Malic acid", "Magnesium", "Total phenols", "Flavanoids", "Nonflavanoid phenols", "Proanthocyanins", "Color intensity", "Hue", "OD280/OD315 of diluted wines", "Proline"]
```

```python
j = 0
plt.figure(figsize=(20,70))
for i in cols_to_be_normalized[0:]:
    
    col_normalized, lamb = sp.stats.boxcox(df[i])
    col_boxcox = ((((df[i])**lamb) - 1) / lamb)

    j+=1
    plt.subplot(13,3,j)
    sns.boxplot(y = col_boxcox,data = df)
    plt.title("Box plot for {}" .format(i))

    j +=1
    plt.subplot(13,3,j)
    sns.histplot(x = col_boxcox, kde=True, data=df)
    plt.title("Histogram for {}" .format(i)) 

    j +=1
    plt.subplot(13,3,j)
    sp.stats.probplot(col_boxcox, plot=plt) 
    plt.title("Probplot for {}" .format(i)) 
    plt.legend(labels = [f'Skew:{col_boxcox.skew()}'], loc = "lower right")

plt.show()
```
