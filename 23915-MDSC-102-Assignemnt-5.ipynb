```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
name = ["class", "Alcohol", "Malic acid", "Ash", "Alcalinity of ash", "Magnesium", "Total phenols", "Flavanoids", "Nonflavanoid phenols", "Proanthocyanins", "Color intensity", "Hue", "OD280/OD315 of diluted wines", "Proline"]
df = pd.read_csv("wine.data", names=name)
df
```

### Boxplot

```python
sns.boxplot(x = 'class', y ='Proline', data=df)
```

```python
def box_plots(df, target, variable, count):
    sns.boxplot(x = target, y = variable,data = df, hue="class")
```

```python
j = 0
plt.figure(figsize=(30,30))
for i in name[1:]:
    j += 1
    plt.subplot(5,3,j)
    box_plots(df, "class", i, j)
    plt.title("Box plot for class and {}" .format(i))

plt.show()
```
